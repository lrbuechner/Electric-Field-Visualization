Not to scale or optimized in any way.

Enjoy the nice plot

- e_0 = 8.8542*10**(-12)
- k = 1/(pi*e_0)
- dist = 10**-10

```python
import math 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

k = 1
mass = 1

def particle(p):
    p = [i for i in p]
    x = p[0] ; y = p[1]
    r = .1
    plt.scatter(x,y,color = 'r' , s=10)
    
    test = []
    circle = np.arange(0,2*math.pi,math.pi*2/(p[2]))
    
    # generating test charges to trace field lines around a (+) 
    for i in circle:
        test.append([x+r*np.cos(i),y+r*np.sin(i)]) #(x,y,vx,vy)
    for i in test:
        plt.scatter(i[0],i[1], color = 'k', s=1)
    return test
```

```python
# setting paramaters
negative = [[0,0,10,.2,'-'],[-.5,.5,10,.7,'-'],[-.5,-.5,10,.7,'-'],[.5,-.5,10,.7,'-'],[.5,.5,10,.7,'-'],
           [1.3,1.3,10,.6,'-'],[1.3,-1.3,10,.6,'-'],[-1.3,-1.3,10,.6,'-'],[-1.3,1.3,10,.6,'-']]

positive = [[-.3,0,16*2,.2,'+'],[.3,0,16*2,.2,'+'],[0,.3,16*2,.2,'+'],[0,-.3,16*2,.2,'+'],[0,-.9,16*2,.4,'+'],
            [0,.9,16*2,.4,'+'],[-.9,0,16*2,.4,'+'],[.9,0,16*2,.4,'+'],[1,1,16*2,.9,'+'],[-1,-1,16*2,.9,'+'],
            [1,-1,16*2,.9,'+'],[-1,1,16*2,.9,'+']]

# initializing plot
plt.figure(figsize = (9,9))
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])

# generating test charges
tracers = []
for i in positive:
    tracers.append(particle(i))
tracers = [j for i in tracers for j in i]
for i in negative:
    plt.scatter(i[0],i[1],color = 'b' , s=10)
    
particles = positive + negative

# utilizing Coulombs Law to iteratively plot field lines 
# sum the components of the force between each charge
step = 15000
while step > 0:
    for i in range(len(tracers)):
        vum_ax = 0
        vum_ay = 0
        for j in range(len(particles)):
            t = tracers[i]
            p = particles[j]
            
            # inward force
            if p[4] == '-':
                x = (t[0]-p[0])
                y = (t[1]-p[1])
                force = (k*p[3]) / (x**2+y**2)
                theta = np.arctan(y/x)
                a_x = force*np.cos(theta) / mass
                a_y = force*np.sin(theta) / mass
                if t[0] > p[0]:
                    vum_ax -= a_x
                else: 
                    vum_ax += a_x
                if t[0] > p[0]:
                    vum_ay -= a_y
                else:
                    vum_ay += a_y
                    
            # outward force
            else: 
                x = (t[0]-p[0])
                y = (t[1]-p[1])
                force = (k*p[3]) / (x**2+y**2)
                theta = np.arctan(y/x)
                a_x = force*np.cos(theta) / mass
                a_y = force*np.sin(theta) / mass
                if t[0] > p[0]:
                    vum_ax += a_x
                else: 
                    vum_ax -= a_x
                if t[0] > p[0]:
                    vum_ay += a_y
                else:
                    vum_ay -= a_y
                        
        # incrimenting position in direction of acceleration 
        tracers[i][0] = tracers[i][0] + vum_ax*.001 
        tracers[i][1] = tracers[i][1] + vum_ay*.001 
        plt.scatter(tracers[i][0],tracers[i][1],color = 'k', s = .2)
        step -= 1
plt.show()
```
